```python
soma = 0

for numero in range(1, 501):
    if numero % 2 != 0 and numero % 3 == 0:
        soma += numero

print("A soma dos ímpares divisíveis por 3 entre 1 e 500 é:", soma)

```

```python

maiores_dezoito = 0
menores_dezoito = 0


for i in range(5):
    idade = int(input(f"Digite a idade da pessoa {i + 1}: "))


    if idade >= 18:
        maiores_dezoito += 1
    else:
        menores_dezoito += 1


print(f"Quantidade de pessoas maiores de 18 anos: {maiores_dezoito}")
print(f"Quantidade de pessoas menores de 18 anos: {menores_dezoito}")
```

```python
def soma_vetores(vetor1, vetor2):
    if len(vetor1) != len(vetor2):
        raise ValueError("Os vetores devem ter o mesmo comprimento para serem somados")

    resultado = []
    for i in range(len(vetor1)):
        resultado.append(vetor1[i] + vetor2[i])

    return resultado

vetor1 = [1, 2, 3]
vetor2 = [4, 5, 6]

resultado = soma_vetores(vetor1, vetor2)
print(resultado)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')


u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)
x = 10 * np.outer(np.cos(u), np.sin(v))
y = 10 * np.outer(np.sin(u), np.sin(v))
z = 10 * np.outer(np.ones(np.size(u)), np.cos(v))


ax.plot_surface(x, y, z, color='b')


ax.set_xlabel('Eixo X')
ax.set_ylabel('Eixo Y')
ax.set_zlabel('Eixo Z')


plt.show()

```
